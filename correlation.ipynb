```python
import pandas as pd
from tqdm import tqdm_notebook as tqdm
from scipy import stats
out=pd.DataFrame()
```

#### Telegram
начало - max(дата первой новости, дата первой цены)

конец - min(дата последней новости, дата последней цены)

```python
df = 0
p = pd.read_csv('C:/Users/79627/WPy64-3741/notebooks/telegram/data/9_join.csv', sep='|')
p['new'] = 1
p['week'] = p.date // 7
p['volatility'] = abs(p.close - p.open) / p.open
p.head(3)
```

```python
out = pd.DataFrame()

for ticker in p.ticker.unique():
    res = dict()
    test = p[p.ticker==ticker][['week', 'volatility', 'new']].groupby(['week', 'volatility']).sum().reset_index()

    res['ticker'] = [ticker]
    res['volatility_pearson_corr'] = [stats.pearsonr(test.volatility, test.new)[0]]
    res['volatility_pearson_p_value'] = [stats.pearsonr(test.volatility, test.new)[1]]

    res['volatility_spearman_corr'] = [stats.spearmanr(test.volatility, test.new)[0]]
    res['volatility_spearman_p_value'] = [stats.spearmanr(test.volatility, test.new)[1]]
    res['volatility_shape'] = [test.shape[0]]


    test = p[p.ticker==ticker][['week', 'vol', 'new']].groupby(['week', 'vol']).sum().reset_index()

    res['volume_pearson_corr'] = [stats.pearsonr(test.vol, test.new)[0]]
    res['volume_pearson_p_value'] = [stats.pearsonr(test.vol, test.new)[1]]

    res['volume_spearman_corr'] = [stats.spearmanr(test.vol, test.new)[0]]
    res['volume_spearman_p_value'] = [stats.spearmanr(test.vol, test.new)[1]]
    res['volume_shape'] = [test.shape[0]]
    
    res = pd.DataFrame.from_dict(res)
    out = out.append(pd.DataFrame(res))
out
```
